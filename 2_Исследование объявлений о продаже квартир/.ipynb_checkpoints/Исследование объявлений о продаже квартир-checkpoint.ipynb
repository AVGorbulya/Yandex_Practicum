### –ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏ –∏—Ö –æ–±—Ä–∞–±–æ—Ç–∫–∞

```python
# —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤ –∫–∞–∫–∏—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö –µ—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ % –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª-–≤–∞ —Å—Ç—Ä–æ–∫
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏ –∏—Ö –¥–æ–ª–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
def print_missing_data(data):
    missing_data = pd.DataFrame({
        'Counts': data.isna().sum(),
        '%_missing': ((data.isna().sum() / len(data)) * 100).round(1)
    })
    missing_data = missing_data.query('Counts != 0')
    print(missing_data)

print_missing_data(data)
```

5.	–î–æ–±–∞–≤–∏–º –∫–æ–ª–æ–Ω–∫—É floor_types —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã –∏ –∑–∞–ø–æ–ª–Ω–∏–º –µ–µ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ ‚Äî ¬´–ø–µ—Ä–≤—ã–π¬ª, ¬´–ø–æ—Å–ª–µ–¥–Ω–∏–π¬ª, ¬´–¥—Ä—É–≥–æ–π¬ª)

```python
def filling_floortyp(floor, floors_total):
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    elif floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        return '–¥—Ä—É–≥–æ–π'
```

```python
df_copy['floor_types'] = df_copy.apply(lambda row: filling_floortyp(row['floor'], row['floors_total']), axis=1)
print(df_copy[['floor', 'floors_total', 'floor_types']].head(15))
```

### –í—ã—è–≤–ª–µ–Ω–∏–µ —Ñ–∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –≤–ª–∏—è—é—Ç —Ü–µ–Ω—É –æ–±—ä–µ–∫—Ç–∞:
    
    –û–ø—Ä–µ–¥–µ–ª–∏–º, –∑–∞–≤–∏—Å–∏—Ç –ª–∏ —Ü–µ–Ω–∞ –æ—Ç —Å–ª–µ–¥—É—é—â–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:
        -	–æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏;
        -	–∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏;
        -	–ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏;
        -	–∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–æ–º–Ω–∞—Ç;
        -	—ç—Ç–∞–∂–∞, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞ (–ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π);
        -	–¥–∞—Ç—ã —Ä–∞–∑–º–µ—â–µ–Ω–∏—è (–¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü, –≥–æ–¥).
        
    –î–ª—è —ç—Ç–æ–≥–æ: 
    - —Å–æ–∑–¥–∞–¥–∏–º –æ—á–∏—â–µ–Ω–Ω—É—é –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ –∫–æ–ø–∏—é –¥–∞—Ç–∞—Ñ—Ä–µ–º–∞,
    - —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏,
    - –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏ —Ä–∞—Å—Å–µ—è–Ω–∏—è. 

1. –î–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –æ—á–∏—â–µ–Ω–Ω—ã–µ –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤.
–°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –æ—á–∏—â–µ–Ω–Ω—ã–π –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤.

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—á–∏—Å—Ç–∫–∏ –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ –∏–∑ –∫–æ–ª–æ–Ω–æ–∫
def remove_excess(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df.loc[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –æ—á–∏—â–µ–Ω–Ω—ã–π –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤
List_to_clean = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'rooms']
df_clean = remove_excess(df_copy, List_to_clean)
```

### –†–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–π —Ü–µ–Ω—ã –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π 

–ó–∞–¥–∞—á–∞: 
–†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π.
 
    - –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π –∏ —Å—Ä–µ–¥–Ω–µ–π —Ü–µ–Ω–æ–π –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –¥–ª—è —ç—Ç–∏—Ö –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤. 
    - –Ω–∞–π—Ç–∏ –∏–∑ —ç—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞.


```python
# 1. —Å—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–æ–π –ø–æ locality_name –∏ –¥–≤—É–º—è –∫–æ–ª–æ–Ω–∫–∞–º–∏ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∏ –º–µ–¥–∏–∞–Ω–æ–π —Ü–µ–Ω–Ω—ã,
# 2. —Å–æ—Ä—Ç–∏—Ä—É–µ–º —Å—Ç–æ–ª–±–µ—Ü  count –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ –æ—Ç–±–∏—Ä–∞–µ–º –ø–µ—Ä–≤—ã–µ 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
# 3. —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ–≤—Ç–æ—Ä–Ω–æ —Å—Ç–æ–ª–±–µ—Ü  median median_price  –ø–æ —É–±—ã–≤–∞–Ω–∏—é,
# 3. –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏

name_group_price = df_copy.groupby('locality_name')\
    .agg(count=('meter_price', 'count'), median_price=('meter_price', 'median'))\
    .sort_values(by='count', ascending=False).head(10)\
    .sort_values(by='median_price', ascending=False)\
    .rename(columns={'count': '–ö–æ–ª-–≤–æ –æ–±—ä—è–≤–ª–µ–Ω–∏–π','median_price': '–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞, —Ç—ã—Å. —Ä—É–±. –∑–∞ –∫–≤.–º'})

name_group_price
```
