# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

# Цель и план работ по проекту "Исследование объявлений о продаже квартир"

    Цель работы:
    провести исследовательский анализ данных, который поможет установить параметры, влияющие на цену объектов.

## План исследования

### Первичное исследование, предобработка и дообогащение данных датафрема:
     1.1 первичный анализ датафрейма методом .info() и .hist(), выявление нелогичных типов данных, аномалий на гистограммах;
     1.2 анализ пропусков в данных, объем, возможные причины и метод заполнения, если это возможно и необходимо для анализа;
     1.3 анализ и исправление типов данных в столбцах при необходимости;
     1.4 проанализируем датафрейм на предмет наличия явных и неявных дубликатов;
     1.5 для проведения дальнейшего анализа дообогатим данные датафрейма следующими колонками:
            •	цена одного квадратного метра;
            •	день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
            •	месяц публикации объявления;
            •	год публикации объявления;
            •	тип этажа квартиры (значения — «‎первый», «последний», «другой»);
            •	расстояние до центра города в километрах.

 ### Исследовательский анализ данных
 
 ####  Исследование отдельных параметров объектов датафрема:
     
      Проведем анализ перечисленных ниже параметров объектов, построим гистограммы и выявим аномалии:
    o	общая площадь;
    o	жилая площадь;
    o	площадь кухни;
    o	цена объекта;
    o	количество комнат;
    o	высота потолков;
    o	тип этажа квартиры («первый», «последний», «другой»);
    o	общее количество этажей в доме;
    o	расстояние до центра города в метрах;
    o	расстояние до ближайшего парка


#### Анализ скорости продажи квартир:
       
       Построим гистограмму, рассчитаем среднее и медиану, ответим на следующие вопросы:
        - сколько обычно занимает продажа; 
        - какие продажи можно считать быстрыми;
        - какие продажи можно считать необычно долгими; 

#### Выявление факторов, которые больше всего влияют цену объекта:
    
    Определим, зависит ли цена от следующих параметров:

### Выявление факторов, которые больше всего влияют цену объекта:
    
    Определим, зависит ли цена от следующих параметров:
        -	общей площади;
        -	жилой площади;
        -	площади кухни;
        -	количества комнат;
        -	этажа, на котором расположена квартира (первый, последний, другой);
        -	даты размещения (день недели, месяц, год).
        
    Для этого: 
    - создадим очищенную от выбросов копию датафрема,
    - рассчитаем коэффициенты корреляции,
    - построим графики рассеяния. 

1. Для корректной оценки необходимо использовать данные очищенные от выбросов.
Создадим датафрейм очищенный от выбросов.

```python
# Функция для очистки от выбросов из колонок
def remove_excess(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df.loc[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
# создадим датафрейм очищенный от выбросов
List_to_clean = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'rooms']
df_clean = remove_excess(df_copy, List_to_clean)
```

2. Определим, зависит ли цена от общей площади

```python
# рассчитаем Коэффициент корреляции
df_clean['last_price'].corr(df_clean['total_area'])
```

```python
#Построим диаграмму рассеяния для last_price и total_area
pivot_total = df_clean.pivot_table(index='total_area', values='last_price', aggfunc='median')
pivot_total.plot(style='o', figsize=(12,5), grid=True, alpha=0.3)
plt.xlabel('Общая площадь, м.кв.')
plt.ylabel('Цена тыс. руб.')
plt.title('Зависимость цены от Общей площади')
plt.show()
```

Вывод.
Коэффициент корреляции 0,67, на графике видна сильная зависимость цены от общей площади.

### Расчет средней цены одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений 

Задача: 
Рассчитать среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений.
 
    - построить сводную таблицу с количеством объявлений и средней ценой квадратного метра для этих населенных пунктов. 
    - найти из этого списка населённые пункты с самой высокой и низкой стоимостью квадратного метра.


```python
# 1. строим сводную таблицу с группировкой по locality_name и двумя колонками с количеством и медианой ценны,
# 2. сортируем столбец  count по убыванию и отбираем первые 10 элементов,
# 3. сортируем повторно столбец  median median_price  по убыванию,
# 3. переименовываем колонки

name_group_price = df_copy.groupby('locality_name')\
    .agg(count=('meter_price', 'count'), median_price=('meter_price', 'median'))\
    .sort_values(by='count', ascending=False).head(10)\
    .sort_values(by='median_price', ascending=False)\
    .rename(columns={'count': 'Кол-во объявлений','median_price': 'Средняя цена, тыс. руб. за кв.м'})

name_group_price
```

```python
# построим график с ценами за кв. м.
name_group_price['Средняя цена, тыс. руб. за кв.м'].plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.xlabel('Средняя цена, тыс. руб. за кв.м')
plt.title('Средняя цена за квадратный метр в населенных пунктах')
plt.gca().invert_yaxis() 
plt.show()
```

Вывод.
    Самая высокая цена в Санкт-Петербурге  - 104 тыс. руб. за кв. м.
    Самая низкая цена в Выборге  - 58 тыс. за кв. м.
