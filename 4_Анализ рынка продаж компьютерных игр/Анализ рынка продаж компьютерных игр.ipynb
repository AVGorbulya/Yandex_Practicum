# Проект " Анализ исторических данных о продажах видеоигр"

## Цель, вводная информация, описание и план исследования

### Цель проекта

        На основе исторических данных о продажах игр, оценок пользователей и экспертов, информации о жанрах и игровых платформах, необходимо выявить определяющие успешность игры закономерности.
        Это исследование должно позволить сделать интернет-магазину «Стримчик», который продаёт по всему миру компьютерные игры, выбор в пользу потенциально популярного продукта и спланировать рекламные кампании.
        То есть, в итоге должны быть ответы на вопросы, 
        в целом и отдельно по регионам мира (Америка, Европа, Азия):
            - какие жанры наиболее популярны?
            - какие платформы наиболее популярны?

### Вводная информация и описание источников данных

В основных данных есть следующая информация:

    •	Name — название игры
    •	Platform — платформа
    •	Year_of_Release — год выпуска
    •	Genre — жанр игры
    •	NA_sales — продажи в Северной Америке (миллионы проданных копий)
    •	EU_sales — продажи в Европе (миллионы проданных копий)
    •	JP_sales — продажи в Японии (миллионы проданных копий)
    •	Other_sales — продажи в других странах (миллионы проданных копий)
    •	Critic_Score — оценка критиков (максимум 100)
    •	User_Score — оценка пользователей (максимум 10)
    •	Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


### План исследования

#### Загрузка и предобработка  данных:
    
    •   Загрузим CSV-файл и изучим общую информацию о датафрейме, выгрузив первые строки.
    •	Приведем названия столбцов к нижнему регистру;
        •	Обработаем пропуски, удалим, если они несущественны, заполним там, где это возможно. 
            -	Объясним алгоритм заполнения пропусков или почему не стали заполнять;
            -	Опишем причины, которые могли привести к пропускам;
            -	Обработаем 'tbd' в столбце с оценкой пользователей.
    •	Проверим и преобразуем данные в нужные типы.
    •	Рассчитаем суммарные продажи в отдельном столбце во всех регионах.

 #### Исследовательский анализ данных:
 
    1.Проанализируем, количество игр, которые выпускалось в разные годы. Определим, важны ли данные за все периоды?
    2.Проанализируем, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. Определим какой характерный срок появляются новые и исчезают старые платформы?
    3.На основании пунктов 1,2 выше, определим актуальный период и сделаем срез данных, который будет использован для дальнейшего анализа. Основной фактор — эти данные должны помочь построить прогноз на 2017 год.
    4.Проанализируем платформы лидируют по продажам, растут или падают. Цель: выбрать несколько потенциально прибыльных платформ.
    5.Построим график «ящик с усами» по глобальным продажам игр в разбивке по платформам. Проанализируем результат.
    6.Проанализируем, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и рассчитаем корреляцию между отзывами и продажами. 
    7.Посмотрим как выводы о корреляции отзывов из п. 6 соотносятся с продажами игр на других платформах.
    8.Проанализируем общее распределение игр по жанрам. Определим какие жанры самые прибыльные и выделяются ли жанры с высокими и низкими продажами.

#### Составим портрет пользователя каждого региона:

      Никакой особой зависимости  между рейтингом  и жанром не обнаружено, рейтинг может быть любой у каждого жанра и в разной пропорции.
      Рейтинги слишком детальные, их можно  сгруппировать более компактно без потери сути рейтинга.
      Есть три основных деления: 
              1. для взрослых. Это рейтинги для взрослых: AO и M. Присвоим этим играм рейтинг  - Adult;

Рассчитаем структуру рейтингов по сгруппированным жанрам в исходном датафрейме

```python
# рассчитаем встречаемость рейтингов  в каждом жанре
genre_rating_group = df.groupby(['genre', 'rating_grouped']).size().reset_index(name='count')
# рассчитаем долю каждого рейтинга в каждом жанре
genre_rating_group['rating_proportion'] = genre_rating_group.groupby('genre')['count'].transform(lambda x: x / x.sum()).multiply(100).round(1)

# сводная таблица со структурой ретингов в каждом  жанре
genre_rating_pivote = genre_rating_group.pivot_table(index='genre', columns='rating_grouped', values='rating_proportion', fill_value=0)

genre_rating_pivote
```

Используем данную статистику для заполнения пропущенных значений в рейтингах.
    
    Алгоритм заполнения следующий:
        - Для каждого жанра пропущенные  значения в rating_grouped заполняем случайным образом в соответствии  с вероятностью его появления в каждом жанре из таблицы genre_rating_pivote.

```python
# фиксируем случайный выбор для повторяемости результатов в дальнейшем
np.random.seed(45)

# заполняем пропуски в колонке rating_grouped
for index, row in df_copy.iterrows():
    genre = row['genre']
    if pd.isnull(row['rating_grouped']): 
        percentages = genre_rating_pivote.loc[genre]  #процентное распределение по жанру
        ratings = ['Adult', 'Everyone', 'Teenager']
        probs = percentages / percentages.sum() #вероятности для каждого рейтинга
        rating = np.random.choice(ratings, p=probs) #рейтинг в соответствии с вероятностями
        df_copy.at[index, 'rating_grouped'] = rating #случайный рейтинг в соответствии с вероятностями

display(df_copy.head(5))
```

### Анализ влияния отзывов пользователей и критиков на продажи внутри одной платформы 

    Возьмем для анализа ранее выбранные как перспективные платформы PS4 и XOne.
    Построим диаграмму рассеяния и рассчитаем корреляцию между отзывами и продажами. 

```python
PS4_sales = df_actual.query('platform == "PS4"')
XOne_sales = df_actual.query('platform == "XOne"')
```

```python
# рассчитаем коэффициент корреляции отзывов критиков и объемов продаж
print('Корреляция с отзывами критиков ', PS4_sales['critic_score'].corr(PS4_sales['total_sales']))
# рассчитаем Коэффициент корреляции отзывов пользователей и объемов продаж
print('Корреляция с отзывами пользователей ',PS4_sales['user_score'].corr(PS4_sales['total_sales']))
```

```python
# рассчитаем Коэффициент корреляции отзывов критиков и объемов продаж
print('Корреляция с отзывами критиков  ', XOne_sales['critic_score'].corr(XOne_sales['total_sales']))
# рассчитаем Коэффициент корреляции отзывов пользователей и объемов продаж
print('Корреляция с отзывами пользователей ',XOne_sales['user_score'].corr(XOne_sales['total_sales']))
```

    Итак, есть умеренная зависимость  продаж с отзывами критиков.
    И наоборот зависимости продаж от отзывов пользователей нет.
    Построим графики рассеяния PS4 и XOne.

```python
#Построим диаграммы зависимости объемов продаж PS4
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# зависимость объемов продаж PS4 от рейтинга критиков
pivot_total_criticPS4 = PS4_sales.pivot_table(index='critic_score', values='total_sales')
pivot_total_criticPS4.plot(ax=axes[0], style='o', grid=True, alpha=1)
axes[0].set_xlabel('Рейтинг критиков')
axes[0].set_ylabel('Продажи, млн. шт.')
axes[0].set_title('Зависимость объемов продаж PS4 от рейтинга критиков')

# зависимость объемов продаж PS4 от рейтинга пользователей
pivot_total_userPS4 = PS4_sales.pivot_table(index='user_score', values='total_sales')
pivot_total_userPS4.plot(ax=axes[1], style='o', grid=True, alpha=1)
axes[1].set_xlabel('Рейтинг пользователей')
axes[1].set_ylabel('Продажи, млн. шт.')
axes[1].set_title('Зависимость объемов продаж PS4 от рейтинга пользователей')

plt.tight_layout()
plt.show()
```

```python
#Построим диаграммы зависимости продаж XOne
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# зависимость объемов продаж XOne от рейтинга критиков
pivot_total_criticXOne = XOne_sales.pivot_table(index='critic_score', values='total_sales')
pivot_total_criticXOne.plot(ax=axes[0], style='o', grid=True, alpha=1)
axes[0].set_xlabel('Рейтинг критиков')
axes[0].set_ylabel('Продажи, млн. шт.')
axes[0].set_title('Зависимость объемов продаж XOne от рейтинга критиков')

# зависимость объемов продаж XOne от рейтинга пользователей
pivot_total_userXOne = XOne_sales.pivot_table(index='user_score', values='total_sales')
pivot_total_userXOne.plot(ax=axes[1], style='o', grid=True, alpha=1)
axes[1].set_xlabel('Рейтинг пользователей')
axes[1].set_ylabel('Продажи, млн. шт.')
axes[1].set_title('Зависимость объемов продаж XOne от рейтинга пользователей')

plt.tight_layout()
plt.show()
```

На графиках видна зависимость объемов продаж от рейтинга критиков, чем рейтинг меньше, тем меньше объемы продаж.

### Cамые популярные платформы (топ-5) регионов NA, EU, JP

Задача: Описать различия между платформами в долях продаж.

```python
# создадим функцию для группировки срезов данных по необходимым для последующего анализа критериям (жанр, платформа итд)
def group_and_agg(data, group_by_column, agg_column):
    # расчитаем общее количество пользователей по каждой группе и медиану исследуемого значения по ним же.
    grouped = data.groupby(group_by_column).agg({agg_column: ['count', 'sum']}).round(1)
    # рассчитаем долю каждой группы в общем количестве в колонке 'struct_%'
    grouped['struct_%'] = (grouped[agg_column]['sum'] /grouped[agg_column]['sum'].sum()).multiply(100).round(1)
    grouped.columns = ['count_sales','sales', 'struct_%']
    grouped = grouped.reset_index().sort_values(by='sales', ascending=False)
    return grouped
```

Рассмотрим есть ли корреляция между рейтингом ESRB и продажами.
Рассчитаем для каждого региона коэффициент корреляции Пирсона, для этого в отдельном столбце присвоим ранг от 1 до 3 каждой группе рейтинга Adult, Teenager, Everyone, так как метод Пирсона принимает на вход только числовые параметры.

```python
# убрали предупреждение Pandas 
pd.options.mode.chained_assignment = None
# создадим и заполним 'rating_range' в соответствии с категориями
df_actual['rating_range'] = np.nan
category_dict = {"Adult": '3', "Teenager": '2', "Everyone": '1'}

for category, value in category_dict.items():
    df_actual.loc[df_actual['rating_grouped'] == category, 'rating_range'] = value

#df_actual['rating_range'] = df_actual['rating_range'].astype(int)
df_actual
```

```python
# расчет коэффициента Пирсона
def correlation(df, sales_column):
    pearson = df[sales_column].corr(df['rating_range']).round(3)
    return pearson
```

```python
print('Корреляция рейтинга ESRB и объемов продаж в регионе NA', correlation(df_actual, 'na_sales'))
print('Корреляция рейтинга ESRB и объемов продаж в регионе EU', correlation(df_actual, 'eu_sales'))
print('Корреляция рейтинга ESRB и объемов продаж в регионе JP', correlation(df_actual, 'jp_sales'))
```

## Проверка отдельных гипотез

Задача:
    Для каждой гипотезы сформулировать нулевую и альтернативную гипотезы.
    Установить пороговое значение alpha.
    Определить критерий для проверки гипотезы.

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

    Для проверки данной гипотезы установим:
        Нулевая гипотеза H0: Средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.
        Альтернативная гипотеза H1: Средний пользовательский рейтинг платформы Xbox One не равен среднему пользовательскому рейтингу платформы PC.
        
    Для проверки гипотез проведем двусторонний t-тест для независимых выборок, так как наши данные подходят для  применения данного теста, а именно:
        
        Условиями применения данного теста являются:
        
        - Генеральные совокупности не зависят друг от друга.
          У нас две независимые выборки оценок пользователей двух разных платформ.
        
        - Выборочные средние нормально распределены.
        
          Это условие также выполняется, так как согласно ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, то выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены близко к нормальному вокруг истинного среднего этой совокупности.
        
        - Дисперсии рассматриваемых генеральных совокупностей равны.
          При этом, если выборки достаточно велики (30 и больше значений) и равны по размеру между собой, то можно оставить параметр equal_var по умолчанию True.
        
        - Уровень значимости для проверки гипотезы поставим на уровне 5%.
        - Тест будет двусторонний, так как альтернативная гипотеза, что средние рейтинги не равны, вне зависимости какой из них больше, а какой меньше.

```python
# задаём уровень значимости
alpha = 0.05

# рейтинги пользователей платформы XOne
XOne_user_score = df_actual.loc[df_actual['platform'] == 'XOne']['user_score']

# рейтинги пользователей платформы PC
PC_user_score = df_actual.loc[df_actual['platform'] == 'PC']['user_score']

# проводим ttest для двух выборок, в качестве альтернативной используем двусторонний тест
results = st.ttest_ind(XOne_user_score, PC_user_score, nan_policy='omit')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

Вывод:  
      
      Не опровергнута гипотеза H0, что средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.
      Полученный результат p-value: 0.5535080500643661 высокий, тест будет выполняться и при более низком значении alpha = 0.01.
      Таким образом, можно сказать, что средние пользовательские рейтинги платформ Xbox One и PC равны.

### Гипотеза 2. Средние пользовательские рейтинги жанров Action  и Sports разные.

 Для проверки данной гипотезы установим:
        
        Нулевая гипотеза H0: Средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.
        Альтернативная гипотеза H1: Средний пользовательский рейтинг жанра Action не равен среднему пользовательскому рейтингу жанра Sports.
        
    Для проверки гипотез проведем двусторонний t-тест для независимых выборок.
    Параметры теста будут такие же как и у предыдущего.

```python
# задаём уровень значимости
alpha = 0.05

# рейтинги пользователей жанра Actione
Action_user_score = df_actual.loc[df_actual['genre'] == 'Action']['user_score']

# рейтинги пользователей платформы PC
Sports_user_score = df_actual.loc[df_actual['genre'] == 'Sports']['user_score']

# проводим ttest для двух выборок, в качестве альтернативной используем двусторонний тест
results = st.ttest_ind(Action_user_score, Sports_user_score, nan_policy='omit')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

Вывод:

      Полученный результат p-value стремиться к нулю.
      Таким образом, можно утверждать, что средние пользовательские рейтинги платформ жанра Action и Sports не равны.
