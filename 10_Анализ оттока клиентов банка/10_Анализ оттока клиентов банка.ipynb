```python
# функция вывода пропущенных значений
# на вход датафрейм
def print_missing_data(data):
    missing_data = pd.DataFrame({
        'Counts':
        data.isna().sum(),
        '%_missing': ((data.isna().sum() / len(data)) * 100).round(1)
    })
    missing_data = missing_data.query('Counts != 0').sort_values(
        by='%_missing', ascending=False)
    print(missing_data)
```

```python
# функция вывода гистограмм с числовыми показателям в колонках
# на вход датафрейм и колонки, по которым не нужно строить гистограммы
def plot_histograms(df, columns_to_drop):
    # удаление стобцов
    df_filtered = df.drop(columns_to_drop, axis=1)
    # отбор числовых данных
    numeric_columns = df_filtered.select_dtypes(include='number').columns
    # подсчет графиков, числа строк и столбцов для subplot
    num_plots = len(numeric_columns)
    num_rows = (num_plots + 1) // 2
    num_cols = 2
    # Создаем графики
    plt.figure(figsize=(12, 20))
    for i, column in enumerate(numeric_columns, 1):
        plt.subplot(num_rows, num_cols, i)
        plt.hist(df_filtered[column])
        plt.title(f'Histogram for {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
#формирует таблицу с основными характеристиками данных в колонках
# на вход датафрейм и колонки исключения
def table_describe(df, columns_to_drop):
    df_filtered = df.drop(columns_to_drop, axis=1)
    numeric_columns = df_filtered.select_dtypes(include='number').columns
    table_D = df_filtered[numeric_columns].describe().round(2)
    display(table_D)
```

```python
# формирует таблицу с расчетом 95 и 99 перцентиля по числовым колонкам датафрейма
# на выход датафрейм и колонки исключения
def table_percentile(df, columns_to_drop):
    df_filtered = df.drop(columns_to_drop, axis=1)
    numeric_columns = df_filtered.select_dtypes(include='number').columns
    # рассчитываем перцентили
    percentile_99 = np.nanpercentile(
        df_filtered[numeric_columns], 99, axis=0).round(0)
    percentile_95 = np.nanpercentile(
        df_filtered[numeric_columns], 95, axis=0).round(0)
    # Создаем DataFrame
    percentile_df = pd.DataFrame(
        percentile_99, index=numeric_columns, columns=['99_percentile'])
    percentile_df['95_percentile'] = percentile_95

    # Вычисляем долю значений, превышающих перцентили
    total_counts = df_filtered[numeric_columns].count()
    values_above_99 = (df_filtered[numeric_columns] > np.nanpercentile(
        df_filtered[numeric_columns], 99, axis=0)).sum()
    values_above_95 = (df_filtered[numeric_columns] > np.nanpercentile(
        df_filtered[numeric_columns], 95, axis=0)).sum()
    percentile_df['99_percentile_%'] = (
        ((values_above_99 / total_counts))*100).round(1)
    percentile_df['95_percentile_%'] = (
        ((values_above_95 / total_counts))*100).round(1)

    display(percentile_df)
```

```python
# функция для строительства heatmap корреляции spearman по одному признаку с удалением ненужных для анализа признаков
# на вход датафрейм, колонка для которой считаем корреляцию, колонки, которые нужно удалить
def plot_heatmap_onecol_spearman(df, column, exclude_columns=[]):
    # отбираем числовые колонки для анализа
    numeric_columns = df.select_dtypes(include='number').columns
    # удаляем колонки, которые необходимо исключить из анализа
    columns_to_analyze = [
        col for col in numeric_columns if col not in exclude_columns]
    # матрица корреляции
    corr_table = df[columns_to_analyze].corr(method='spearman')
    # тепловая карта
    plt.figure(figsize=(3, 6))
    sns.heatmap(corr_table.loc[:, [column]].drop(column, axis=0).sort_values(column, ascending=True),
                annot=True, fmt='.0%', cmap='vlag', center=0)
    plt.title(f'Корреляция признаков с {column}', fontsize=8)
    plt.show()
```

```python
# функция для строительства heatmap корреляции Фай-Фи по одному признаку с удалением ненужных для анализа признаков
# на вход датафрейм, колонка для которой считаем корреляцию, колонки, которые нужно удалить
# не показывает направление корреляции, поэтому когда важно  определить направление используем plot_heatmap_onecol_spearman
def plot_heatmap_onecol_phik(df, column, exclude_columns=[]):
    # отбираем числовые колонки для анализа
    interval_cols = df.select_dtypes(include='number').columns
    # удаляем колонки, которые необходимо исключить из анализа
    columns_to_analyze = [
        col for col in interval_cols if col not in exclude_columns]
    # матрица корреляции
    phik_matrix = df[columns_to_analyze].phik_matrix(
        interval_cols=interval_cols)
    # тепловая карта
    plt.figure(figsize=(3, 6))
    sns.heatmap(phik_matrix.loc[:, [column]].drop(column, axis=0).sort_values(column, ascending=False),
                annot=True, fmt='.0%', cmap='coolwarm', center=0)
    plt.title(f'Корреляция признаков с {column}', fontsize=8)
    plt.show()
```

```python
# Проверка равенства дисперсий между двумя выборками с использованием теста Левена.
def test_levene(sample1, sample2, alpha=0.05):
    statistic, p_value = levene(sample1, sample2)
    print('p-значение: ', "{0:.5f}".format(p_value))
    if p_value > alpha:
        print("Нет оснований отвергнуть нулевую гипотезу о равенстве дисперсий.")
    else:
        print("Отвергаем нулевую гипотезу. Дисперсии в выборках различаются статистически значимо.")
```

```python
#Проверка выборки на нормальность распределения с использованием теста Шапиро-Уилка.
def test_shapiro (sample, alpha=0.05):
    results = st.shapiro(sample)
    print('p-значение: ', "{0:.5f}".format(results.pvalue))

    if results.pvalue > alpha:
        print('Нет оснований отвергнуть нулевую гипотезу. Распределение выборки нормальное.')
    else:
        print('Отвергаем нулевую гипотезу. Распределение выборки не нормальное')
```

```python
# t-test Стьюдента для независимых выборок
def t_test_ind(sample1, sample2, alternative, alpha=0.05):

    results = st.ttest_ind(
        sample1, sample2, equal_var=False, alternative=alternative)

    print('p-значение: ', "{0:.5f}".format(results.pvalue))

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу: Средние значения двух выборок статистически значимо отличаются.')
    else:
        print('Нет оснований отвергнуть нулевую гипотезу: Статистически значимых отличий между средними значениями двух выборок нет.')
```

```python
# тест Манна-Уитни
def mann_whitneyu_test(sample1, sample2, alternative, alpha=0.05):

    results = st.mannwhitneyu(sample1, sample2, True, alternative=alternative)

    print('p-значение:', "{0:.5f}".format(results.pvalue))

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу: Средние значения двух выборок статистически значимо отличаются.')
    else:
        print('Нет оснований отвергнуть нулевую гипотезу: Статистически значимых отличий между средними значениями двух выборок нет.')
```

#### Посмотрим на сводный портрет  отточного и неотточного клиента

```python
category_group = df_copy.groupby('churn').agg({
          
        'score':'mean',
        'age':'mean',
        'equity':'median',
        'products':'median',
        'credit_card':lambda x: x.value_counts().index[0],
        'last_activity':lambda x: x.value_counts().index[0],
        'balance':'mean',
        'est_salary':'mean',
        'gender':lambda x: x.value_counts().index[0],
        'city':lambda x: x.value_counts().index[0]
    }).round(0)
# средние значения для 'balance' и 'est_salary'
category_group['balance_median'] = df_copy.groupby('churn')['balance'].median().round(0)
category_group['est_salary_median'] = df_copy.groupby('churn')['est_salary'].median().round(0)

#category_group = category_group.T
category_group = category_group.T.rename(columns={0:'NotChurn', 1:'Churn'})
for index, row in category_group.iterrows():
    try:
        category_group.at[index, 'diff_%'] = round(((row['Churn'] - row['NotChurn']) / row['NotChurn']) * 100, 0)
    except ZeroDivisionError:
        category_group.at[index, 'diff_%'] = np.nan
    except TypeError:
        category_group.at[index, 'diff_%'] = np.nan

category_group
```

###  Проверка статистических гипотез.

#### Гипотеза о различии дохода между клиентами, которые ушли и клиентами, которые остались.

    Сформулируем гипотезы:
            
            Нулевая гипотеза H0: средний доход отточного клиента равен среднему доходу клиентов, оставшихся с банком.

            Альтернативная гипотеза H1: средний доход отточного клиента не равен среднему доходу клиентов, оставшихся с банком.

Проверка t-критерием Стьюдента

 Условиями применения t-критерием Стьюдента являются:
        
        - Генеральные совокупности не зависят друг от друга.
            у нас две независимые выборки отточных и неотточных клиентов.
        - Выборочные средние нормально распределены.
            Для проверки этого условия протестируем обе выборки на нормальность методом Шапиро-Уилка.
        - Дисперсии рассматриваемых генеральных совокупностей равны.
            Рассчитаем и сравним стандартные отклонения данных выборок и протестируем значимость тестом Левена.

1. Сформируем выборки и рассчитаем стандартные отклонения и относительную разницу в них

```python
Salary_Churn = df_copy.query('churn == 1')['est_salary']
Salary_NotChurn = df_copy.query('churn == 0')['est_salary']
print('Стандартное отклонение средней зарплаты отточных клиентов:', Salary_Churn.std().round())
print()
print('Стандартное отклонение средней зарплаты неотточных клиентов:', Salary_NotChurn.std().round())
print()
print('Относительная разница стандартных отклонений, в %:', ((Salary_Churn.std().round()-Salary_NotChurn.std())/Salary_NotChurn.std()*100).round(0))
```

Дополнительно убедимся в значимости различий дисперсий выборок проведя тест Левена

```python
test_levene(Salary_Churn, Salary_NotChurn, alpha=0.05)
```

    Таким образом, дисперсии выборок отличаются статистически значимо, поэтому при проведении t-теста Стьюдента установим параметр equal_var=False. Это означает, что тест будет учитывать различия в дисперсиях при расчете p-значений.

2. Проверим выборки на нормальность методом Шапиро-Уилка

2.1 Проверим неотточных клиентов

```python
test_shapiro(Salary_NotChurn, alpha=0.05)
```

2.2 Проверим отточных клиентов

```python
test_shapiro(Salary_Churn, alpha=0.05)
```

     Итак обе выборки распределены не нормально, имеются существенные выбросы.
    При таких данных для тестирования гипотезы о равенстве среднего дохода отточных и неотточных клиентов нужно  использовать тест Манна-Уитни, а не t-тест Стьюдента.
    Проведем оба теста и проверим есть ли разница в результатах.

3. Проведем двусторонний t-тест Стьюдента для независимых выборок

```python
t_test_ind (Salary_Churn, Salary_NotChurn, alpha = 0.05, alternative='two-sided')
```

T-тест Стьюдента не обнаружил статистически значимых различий в среднем доходе отточного и неотточного клиента.

4. Проведем тест на основе непараметрического теста Манна-Уитни.

    Главное отличие непараметрических методов в том, что они работают с рангами — номерами значений в упорядоченном ряду, никак не учитывая сами значения. Поэтому к ним прибегают тогда, когда работа с самими значениями невозможна из-за выбросов, сильно сдвигающих параметрические результаты.

```python
mann_whitneyu_test(Salary_Churn, Salary_NotChurn, alpha = 0.05, alternative='two-sided')
```

    Вывод:
        Тест Стьюдента не обнаружил статистически значимых различий в уровне дохода отточных и неотточных клиентов. Тест применялся на выборках с ненормальным распределением и выбросами, что противоречит  критериям для применения данного теста.
        Тест Манна-Уитни, непараметрический показал наличие статистически значимых различий в доходах отточных и неотточных клиентов. Данный тест устойчив к выбросам, поэтому в данном случае его результаты более достоверны.
    
    Вывод, доходы отточных клиентов отличаются от доходов неотточных клиентов, доход отточных клиентов выше (тест с параметром  alternative=greater также отвергает нулевую гипотезу о равенстве средних и подтверждает, что доход отточных клиентов выше, чем у неотточных, что согласуется портретом отточного клиента, где медианная зарплата отточного клиента была выше на 6%.
